```python
import numpy as np
import pandas as pd  #dataframes
import re   #text search similar to linux
from sklearn.feature_extraction.text import TfidfVectorizer #text into features (number)
from nltk.corpus import stopwords #natural lang toolkit useless words
from nltk.stem.porter import PorterStemmer  #takes a word to remove and suffix and takes root word only
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#sk learn provides numerous ML algo's
from sklearn.metrics import accuracy_score;

import joblib
```

```python
import nltk
nltk.download('stopwords')
```

```python
#print(stopwords.words('english'))
#these words dont add much value
```

DATA PREPROCESS

```python
news_dataset = pd.read_csv('/content/drive/MyDrive/FAKE NEWS (IIT JAMMU)/train.csv')
news_dataset.head()
```

```python
#data pre process missing values
news_dataset.isnull().sum()
#data cleaning remove null values
# null values with null string
news_dataset=news_dataset.fillna('')
```

```python
news_dataset.isnull().sum()
```

```python
# we will predict using title and author name
#combine title with author in new column content
news_dataset['content']=news_dataset['author']+' '+news_dataset['title']
#now we will use content column + labels to predict news
```

```python
#drop column 'label' from dataset
lab=news_dataset.drop(columns='label',axis=1)

#store the column 'label' to another variable
y=news_dataset['label']

```

Stemming: reduce a word to root word

example: actor,actress,acting --> root word is 'act'

```python
port_stem = PorterStemmer()   #load stemmer into var
```

```python
#content is input

def stemming(content):
  #below line remove all characters which are not alphabets, replce no. commas with space!
  stemmed_content = re.sub( '[^a-z,A-z]',' ', content)
  stemmed_content=stemmed_content.lower()

  #below line convert each word into a list!
  stemmed_content=stemmed_content.split()
  stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content=' '.join(stemmed_content)
  return stemmed_content
```

```python
#apply functioon to content column
news_dataset['content']=news_dataset['content'].apply(stemming)
```

```python
news_dataset['content'].head()
#we removed stop words and kept root words only
```

```python
#seperate data and label
X=news_dataset['content'].values
Y=news_dataset['label'].values
```

```python
#convert text  into numerical data
vector=TfidfVectorizer() #term frequency vectorizer, assign number to repeating important words
vector.fit(X)  #convert text to numbers

X=vector.transform(X)
```

```python
print(X)
```

Splitt train and test data 80/20

```python
X_train , X_test, Y_train , Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

Training the Logistic regression model

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)

```

```python
#save model
joblib_file = "logistic_regression_model.pkl"
joblib.dump(model, joblib_file)
print(f"Model saved as {joblib_file}")
```

```python
#model evaluation accuracy score on training data
X_train_predict=model.predict(X_train)
train_data_accuracy= accuracy_score(X_train_predict,Y_train)
print(train_data_accuracy)
```

```python
X_test_predict=model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_predict,Y_test)
print(test_data_accuracy)
```

```python
loaded_model = joblib.load(joblib_file)
```

Predicting UI

```python



def predict_news(input_text):
    processed_text = stemming(input_text)
    vectorized_text = vector.transform([processed_text])
    prediction = loaded_model.predict(vectorized_text)
    return 'Real' if prediction == 0 else 'Fake'
stsr=input("Enter the news: ")
print(predict_news(stsr))
```

```python
#orediction based on new data
X_new=X_test[0] #first row test
predict=model.predict(X_new)
print(predict)
if(predict==0):
  print("real")
else:
  print("fake")
print("real label: ",end=" ")
print(Y_test[0])
```
